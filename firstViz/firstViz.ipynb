```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
def getDivisors(n):
    divisors = list()
    for i in range(1,n+1):
        if(n%i==0):
            divisors.append(i)
    return divisors

divisorsA = getDivisors(37)
divisorsB = getDivisors(75)
fig, ax = plt.subplots()
venn2([set(divisorsA), set(divisorsB)],set_labels = ('divisores de 37', 'divisores de 75'))
fig.savefig("venn.png")
plt.show()
plt.clf()
plt.close()
```

```python
size_of_groups=[75,37]
names=['75','37']
fig, ax = plt.subplots()
# Create a pieplot
plt.pie(size_of_groups,labels=names, colors=['red','green'])
# add a circle at the center
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
fig.savefig("donut.png")
plt.show()
plt.clf()
plt.close()
```

```python
import numpy as np
fig, ax = plt.subplots()
height = [37,75]
bars = ('A', 'B')
y_pos = [0,1]
plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
fig.savefig("bars.png")
plt.show()
plt.clf()
plt.close()
```

```python
import pandas as pd
import networkx as nx
fig, ax = plt.subplots()
# Build a dataframe with 4 connections
df = pd.DataFrame({ 'from':([37]*len(divisorsA) + [75]*len(divisorsB)), 'to':(divisorsA + divisorsB)})
df 
# Build your graph
G=nx.from_pandas_edgelist(df, 'from', 'to')
 
# Plot it
nx.draw(G, with_labels=True)
fig.savefig("graph.png")
plt.show()
```
